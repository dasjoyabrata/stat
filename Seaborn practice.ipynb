```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(10)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
```

```python
sns.pairplot(tips,hue="sex")
```

```python
import matplotlib.pyplot as plt
fig= plt.figure()
ax= fig.add_axes([0,0,2,2])
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=True)
```

```python
import matplotlib.pyplot as plt
fig= plt.figure()
ax= fig.add_axes([0,0,2,2])
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
import matplotlib.pyplot as plt
fig= plt.figure()
ax= fig.add_axes([0,0,2,2])
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
import matplotlib.pyplot as plt
fig= plt.figure()
ax= fig.add_axes([0,0,2,2])
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
import seaborn as sns

tips= sns.load_dataset("tips")
```

```python
flights = sns.load_dataset('flights')
```

```python
tips.head(10)
```

```python
flights.head(10)
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm')
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
import seaborn as sns
sns.load_dataset('iris')
irish.head(10)
```

```python
ic=irish.corr()
```

```python
sns.heatmap(ic,cmap='coolwarm',annot=True)
```

```python
sns.pairplot(irish)
```

```python
pg=sns.PairGrid(irish)
pg.map_diag(sns.distplot)
pg.map_upper(sns.kdeplot)
pg.map_lower(sns.scatterplot)
```
